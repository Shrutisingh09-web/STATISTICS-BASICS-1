Question 5: Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
(Include your Python code and output in the code box below.)
Answer:


Note: The dataset actually has two modes (12 and 24) because both occur 3 times. However, Python's mode() function returns the first mode it encounters, which is 12. For academic submissions, you may mention that the distribution is bimodal with modes 12 and 24.


```python
# Python program to compute Mean, Median, and Mode

from statistics import mean, median, mode

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate Mean
mean_value = mean(numbers)

# Calculate Median
median_value = median(numbers)

# Calculate Mode
mode_value = mode(numbers)

# Display results
print("Mean =", mean_value)
print("Median =", median_value)
print("Mode =", mode_value)
```

Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
(Include your Python code and output in the code box below.)
Answer

Interpretation:
The covariance (287.5) is positive, indicating that both variables tend to increase together.
The correlation coefficient (0.9946) is very close to +1, showing a very strong positive linear relationship between list_x and list_y.

```python
# Python program to compute Covariance and Correlation Coefficient

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Compute Covariance Matrix
cov_matrix = np.cov(list_x, list_y)

# Extract Covariance
covariance = cov_matrix[0, 1]

# Compute Correlation Coefficient
correlation = np.corrcoef(list_x, list_y)[0, 1]

# Display Results
print("Covariance =", covariance)
print("Correlation Coefficient =", correlation)
```

Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
(Include your Python code and output in the code box below.)
Answer:

Explanation:

A boxplot is a graphical representation of data that shows the median, quartiles, and potential outliers. For the given dataset:

First Quartile (Q1) = 15.75
Third Quartile (Q3) = 23.25
Interquartile Range (IQR) = 7.5
Lower Bound = 4.5
Upper Bound = 34.5

Any value below 4.5 or above 34.5 is considered an outlier. Since 35 is greater than the upper bound of 34.5, it is identified as an outlier.

Therefore, the dataset contains one outlier: 35. The boxplot will display this value as a separate point beyond the upper whisker, indicating that it lies unusually far from the rest of the data.

```python
# Python script to draw a boxplot and identify outliers

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Calculate Quartiles
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Determine Outlier Limits
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify Outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
print("Outliers =", outliers)

# Draw Boxplot
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.show()
```

Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

Answer:

As a data analyst, I would use covariance and correlation to determine whether there is a relationship between advertising spend and daily sales.

Covariance measures the direction of the relationship between two variables. If the covariance is positive, it indicates that both variables tend to increase together. If it is negative, one variable tends to increase while the other decreases. However, covariance does not indicate the strength of the relationship.

Correlation measures both the direction and strength of the relationship between two variables. The correlation coefficient ranges from -1 to +1:

+1 indicates a perfect positive relationship.
0 indicates no relationship.
-1 indicates a perfect negative relationship.

In this case, if advertising spend increases and daily sales also increase, we would expect a positive covariance and a correlation coefficient close to +1, indicating a strong positive relationship.

Interpretation

The correlation coefficient is approximately 0.99, which is very close to +1. This indicates a very strong positive relationship between advertising spend and daily sales. As advertising expenditure increases, daily sales also tend to increase significantly. Therefore, the marketing team can conclude that higher advertising spending is strongly associated with higher sales, although further analysis would be needed to establish causation.

```python
# Python program to compute correlation between advertising spend and daily sales

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Calculate Correlation Coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Correlation Coefficient =", correlation)
```

Question 9: Your team has collected customer satisfaction survey data on a scale of 1-
10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

Answer:

Before launching a new product, it is important to understand the distribution of customer satisfaction scores. Several summary statistics and visualizations can be used for this purpose.

Summary Statistics
Mean (Average):
Measures the overall average satisfaction score.
Helps determine the general level of customer satisfaction.
Median:
Represents the middle value of the dataset.
Useful when there are extreme scores that may affect the mean.
Mode:
Indicates the most frequently occurring satisfaction score.
Helps identify the most common customer opinion.
Standard Deviation:
Measures how spread out the scores are around the mean.
A low standard deviation indicates that responses are clustered near the average, while a high standard deviation indicates greater variability.
Minimum and Maximum Values:
Show the lowest and highest satisfaction scores received.
Visualizations
Histogram:
Displays the frequency distribution of survey scores.
Helps identify patterns such as concentration, spread, and skewness.
Boxplot:
Shows the median, quartiles, and potential outliers.
Useful for detecting unusual responses.
Bar Chart:
Displays the frequency of each satisfaction score.

Among these, a histogram is particularly useful because it provides a clear visual representation of how customer satisfaction scores are distributed.

Interpretation

The histogram shows that most customer satisfaction scores are concentrated between 7 and 9, indicating generally positive customer feedback. The score 7 occurs most frequently, suggesting that customers are largely satisfied. There are very few low ratings (4 and 5), and only one perfect rating (10). Overall, the distribution suggests a favorable customer response, which may indicate a positive outlook for the new product launch.

```python
# Python program to create a histogram of survey scores

import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create Histogram
plt.hist(survey_scores, bins=7)

plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")

plt.show()
```
